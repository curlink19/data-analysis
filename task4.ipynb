Изучите код, в котором скачиваются страницы сайта `simple.wikipedia.org` с <a href="https://mipt-stats.gitlab.io/courses/ad_fivt/data_parsing.html">занятия про парсинг данных</a> и **скопируйте его в решение данной задачи**. Код, реализующий обход в ширину с фильтрацией и нормализацией урлов, приведён в виде функции.

```python
from urllib.parse import urlparse, urldefrag, urljoin
from urllib.request import urlopen
from bs4 import BeautifulSoup
from queue import Queue
import time

# Сделаем новое начало
SEED = "https://simple.wikipedia.org/wiki/Chess"

def download_from_the_internet(url):
    try:
        return urlopen(url).read().decode('utf-8')
    except KeyboardInterrupt:
        raise
    except:
        return None

    
def extract_links_from_html(url, html):
    parser = BeautifulSoup(html)
    # Формируем ссылки на те страницы, на которые ссылается документ
    return [urljoin(url, link.get('href')) for link in parser.findAll('a')]


def extract_text_info_from_html(html):
    soup = BeautifulSoup(html, features="html.parser")
    for script in soup(["script", "style"]):
        script.extract()
    
    # Объединяем строки текста
    text = soup.get_text()
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = '\n'.join(chunk for chunk in chunks if chunk)

    # Находим название на странице
    title = soup.find('title').string
    
    return {'text': text, 'title': title}
    

def is_wiki_article(url):
    if 'simple.wikipedia' not in url:
        return False
    
    prohibited = ['Category', 'Special', 'Wikipedia',
                  'php', 'Help']
    
    # Проверяем, есть ли запрещённые строки в ссылке
    for token in prohibited:
        if token in url:
            return False
        
    return True
```

Код обхода в ширину:

```python
# Сделаем более умную нормализацию ссылки
def smart_normalize(url):
    url = url_normalize(url)
    prohibited = ['#', '?']
    for badCharacter in prohibited:
        position = url.find(badCharacter)
        if position != -1:
            url = url[0:position]
    return url

def load_web_pages(seed, max_downloads, filtration_function):
    # Создаём список со страницами
    pages_json = []
    
    # Создаём очередь для обхода в ширину
    q = Queue()
    q.put(seed)

    already_visited = set()
    n_downloads = 0
    time_start = time.time()

    while not q.empty():
        # Нормализуем урл
        main_url = smart_normalize(q.get())
        if main_url in already_visited:
            continue
        already_visited.add(main_url)
        html = download_from_the_internet(main_url)

        # Извлекаем ссылки из страницы
        children_links = extract_links_from_html(main_url, html)
        time.sleep(1)

        # Извлекаем текст страницы
        text_info = extract_text_info_from_html(html) 

        # Добавляем запись в таблицу
        text_info['url'] = main_url
        text_info['childrenLinksNumber'] = len(children_links)
        pages_json.append(text_info)

        # Выводим информацию о скачивании
        if n_downloads % 20 == 0:
            print(n_downloads / max_downloads)

        n_downloads += 1
        if n_downloads > max_downloads:
            break

        # Добавляем ещё не посещённые ссылки в очередь
        for child in children_links:
            if smart_normalize(child) not in already_visited \
            and filtration_function(child):
                q.put(child)
                
    return pages_json
```

```python
# Будем определять принадлежность страницы по вхождениям слов
def is_about_chess(text):
    indicatorPrefixes = ['chess', 'grandmaster', 'fide']
    text = text.split()
    text = [word.lower() for word in text]
    for word in text:
        for prefix in indicatorPrefixes:
            if word.startswith(prefix):
                return 1
    return 0

# Созданим новую колонку с информацией о том, подходит ли страница
pages_df['isAboutChess'] = pd.Series(index=pages_df.index, dtype=int)
for position in pages_df.index:
    pages_df['isAboutChess'].iloc[position] = is_about_chess(pages_df['text'].iloc[position])

counter = 0
for position in pages_df.index:
    counter += pages_df['isAboutChess'].iloc[position]

print(counter / len(pages_df.index))
```

Скачайте снова не менее $500$ страниц, но уже с новой функцией фильтрации. Добавьте следующую информацию к каждой странице:

* глубина в обходе,
* время скачивания страницы.

Для замера времени можете воспользоваться функцией `time` из пакета `time`.

```python
def is_selected_page(url):
    url = smart_normalize(url)
    if is_wiki_article(url) == False:
        return False
    if 'file' in url.lower():
        return False
    return True

# Сделаем более умную нормализацию ссылки
def smart_normalize(url):
    url = url_normalize(url)
    prohibited = ['#', '?']
    for badCharacter in prohibited:
        position = url.find(badCharacter)
        if position != -1:
            url = url[0:position]
    return url

def load_web_pages(seed, max_downloads, filtration_function):
    # Создаём список со страницами
    pages_json = []
    
    # Глубина входа 
    depth = dict()
    
    # Создаём очередь для обхода в ширину
    q = Queue()
    q.put(seed)
    
    depth[smart_normalize(seed)] = 0

    already_visited = set()
    n_downloads = 0
    time_start = time.time()

    while not q.empty():
        # Нормализуем урл
        main_url = smart_normalize(q.get())
        if main_url in already_visited:
            continue
        already_visited.add(main_url)
        current = time.time()
        html = download_from_the_internet(main_url)
        current = time.time() - current

        # Извлекаем ссылки из страницы
        children_links = extract_links_from_html(main_url, html)
        time.sleep(1)

        # Извлекаем текст страницы
        text_info = extract_text_info_from_html(html) 

        # Добавляем запись в таблицу
        text_info['url'] = main_url
        text_info['childrenLinksNumber'] = len(children_links)
        text_info['time'] = current
        text_info['depth'] = depth[main_url]
        pages_json.append(text_info)

        # Выводим информацию о скачивании
        if n_downloads % 5 == 0:
            print(n_downloads / max_downloads)

        n_downloads += 1
        if n_downloads > max_downloads:
            break

        # Добавляем ещё не посещённые ссылки в очередь
        for child in children_links:
            if smart_normalize(child) not in already_visited \
            and filtration_function(child):
                depth[smart_normalize(child)] = depth[smart_normalize(main_url)] + 1
                q.put(child)
                
    return pages_json

# Посчитаем теперь новым способом 
pages_df_smart = pd.DataFrame(load_web_pages(SEED, 500, is_selected_page))
print(pages_df_smart.head(5))
```
